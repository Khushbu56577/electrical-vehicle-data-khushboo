##Problem Statement/Objective
  The adoption of Electric Vehicles (EVs) is increasing globally due to environmental concerns and advancements in technology. To maximize their efficiency, it's crucial to analyze data collected from EVs and charging stations. This data includes usage patterns, charging behaviors, energy consumption, and environmental impact. Analyzing this data can provide insights into optimizing EV performance, improving charging infrastructure, predicting energy demands, and enhancing user experience.


  Data Description

    • VIN (1-10): Partial Vehicle Identification Number.
    • County: The county in which the vehicle is registered.
    • City: The city in which the vehicle is registered.
    • State: The state in which the vehicle is registered. It appears that this dataset may be focused on Washington (WA) state.
    • Postal Code: The postal code where the vehicle is registered.
    • Model Year: The year of the vehicle model
    • Make: The manufacturer of the vehicle.
    • Model: The model of the vehicle.
    • Electric Vehicle Type: The type of electric vehicle, e.g., Battery Electric Vehicle (BEV).
    • Clean Alternative Fuel Vehicle (CAFV) Eligibility: Eligibility status for clean alternative fuel vehicle programs.
    • Electric Range: The maximum range of the vehicle on a single charge (in miles).
    • Base MSRP: The Manufacturer’s Suggested Retail Price.
    • Legislative District: The legislative district where the vehicle is registered.
    • DOL Vehicle ID: Department of Licensing Vehicle Identification.
    • Vehicle Location: Geographic coordinates of the vehicle location.
    • Electric Utility: The electric utility service provider for the vehicle’s location.
    • 2020 Census Tract: The census tract for the vehicle’s location.

##Descriptive Statistics:

###What are the mean, median, and standard deviation of the base MSRP for the vehicles in the dataset?

```python
# Calculate mean, median, and standard deviation of Base MSRP
mean_msrp = data['Base MSRP'].mean()
median_msrp = data['Base MSRP'].median()
std_dev_msrp = data['Base MSRP'].std()

print(f"Mean Base MSRP: {mean_msrp}")
print(f"Median Base MSRP: {median_msrp}")
print(f"Standard Deviation of Base MSRP: {std_dev_msrp}")
```

###Data Distribution:

###What is the distribution of vehicle makes in the dataset? Represent it using a bar chart.

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Make',data=data)
```

##Model Year Analysis:

###What are the most common model years in the dataset? Provide a frequency table and histogram.

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Model Year',data=data)
```

##Electric Vehicle Type:

###What is the proportion of Battery Electric Vehicles (BEV) versus other types of electric vehicles

```python
bev_count = data[data['Electric Vehicle Type'] == 'Battery Electric Vehicle (BEV)'].shape[0]
total_count = data.shape[0]

proportion_bev = bev_count / total_count

print(f"Proportion of BEVs: {proportion_bev:.2%}")
```

##Electric Range Analysis:

###What is the average electric range for vehicles of different makes? Provide a summary table.

```python
data['Electric Range'] = pd.to_numeric(data['Electric Range'], errors='coerce')

avg_range_by_make = data.groupby('Make')['Electric Range'].mean().reset_index()

print("Average Electric Range by Makes:")
print(avg_range_by_make)
```

##County Distribution:

###How are vehicles distributed across different counties in Washington state? Represent the distribution using a pie chart.

```python
county_counts = data['County'].value_counts()

plt.figure(figsize=(10, 8))
county_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Vehicles Across Counties in Washington State')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

##Price Analysis:

###Compare the average base MSRP of vehicles eligible for the Clean Alternative Fuel Vehicle (CAFV) program versus those that are not.

```python
avg_msrp_cafv_eligible = data[data['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'Eligible']['Base MSRP'].mean()
avg_msrp_not_cafv_eligible = data[data['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'Not eligible']['Base MSRP'].mean()

print(f"Average Base MSRP for CAFV Eligible Vehicles: ${avg_msrp_cafv_eligible:.2f}")
print(f"Average Base MSRP for Non-CAFV Eligible Vehicles: ${avg_msrp_not_cafv_eligible:.2f}")
```

##Geographical Analysis:

###How does the base MSRP vary across different cities in Washington state?

```python
avg_msrp_by_city = data.groupby('City')['Base MSRP'].mean().reset_index()

print("Average Base MSRP Across Different Cities in Washington State:")
print(avg_msrp_by_city)
```

##Legislative Districts:

##Electric Range Correlation:

###Is there a correlation between the electric range and the base MSRP of the vehicles? Provide the correlation coefficient and interpret the result.

```python
correlation_coefficient = data['Electric Range'].corr(data['Base MSRP'])

print(f"Correlation Coefficient between Electric Range and Base MSRP: {correlation_coefficient:.2f}")

if correlation_coefficient > 0:
    print("There is a positive correlation between electric range and base MSRP.")
elif correlation_coefficient < 0:
    print("There is a negative correlation between electric range and base MSRP.")
else:
    print("There is no linear correlation between electric range and base MSRP.")
```

##VIN Analysis:

###Identify any patterns or commonalities in the VIN (1-10) for the vehicles. Are there any frequent prefixes or suffixes?

```python
data['VIN (1-10)'] = data['VIN (1-10)'].str.replace(' ', '')
vin_prefix_counts = data['VIN (1-10)'].value_counts()

# Identify frequent prefixes
frequent_vin_prefixes = vin_prefix_counts[vin_prefix_counts > 1]

print("Frequent VIN Prefixes:")
print(frequent_vin_prefixes)
```

##Eligibility Status:

###What percentage of vehicles are eligible for the Clean Alternative Fuel Vehicle (CAFV) program?

```python
eligible_count = data[data['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'Eligible'].shape[0]
total_count = data.shape[0]

percentage_eligible = (eligible_count / total_count) * 100

print(f"Percentage of Vehicles Eligible for CAFV Program: {percentage_eligible:.2f}%")
```

##Model Year Trend:

###Analyze the trend in the number of registered electric vehicles by model year. Provide a line chart to show any increase or decrease over the years.

```python
model_year_counts = data['Model Year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(model_year_counts.index, model_year_counts.values, marker='o', linestyle='-')
plt.title('Trend in Number of Registered Electric Vehicles by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##Range vs. Year:
